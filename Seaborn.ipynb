```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mtcars = pd.read_csv("mtcars.csv")
mtcars
```

**Bar Plot**

```python
res = sns.barplot(mtcars["cyl"],mtcars["mpg"],Color = '#FF7F0E')
res
```

```python
res = sns.barplot(x = "cyl" , y = "mpg", data = mtcars,palette ="Set3")
res
```

**Countplot - to count the number of values**
 



```python
res = sns.countplot(x = "cyl", data = mtcars, palette ="Set1")
```

```python
res = sns.countplot(x = "carb", data =mtcars, palette = "rocket")
```

```python
res = sns.countplot(y ="mpg", data =mtcars, palette ="Set2")
```

**Countplot with Hue parameter [Hue is nothing but group by function]**

```python
sns.countplot(x="gear", hue = "carb", data = mtcars, palette="Set1")  
```

```python
hist = sns.histplot(mtcars.hp, bins = 10, color ="#16FF32")
hist
```

# **EDA Graph**
**Used to visualize the strength of association between the attributes given in the dataset. Technically speaking it gives the correlation between the attributes.**

**Note: Darker the color of the heatmap indicates the heatmap is negative whereas Lighter the color indicates the heatmap is positive**

```python
sns.heatmap(mtcars.corr(), cbar = True, linewidth = 0.5)
```

**Iris Dataset**

```python
iris = sns.load_dataset("iris")
iris
```

```python
iris["species"].value_counts()
```

```python
sns.scatterplot(x="sepal_length", y = "petal_length", data = iris)
```

```python
sns.scatterplot(x="sepal_length", y = "petal_length", data = iris, hue = "species")
```

```python
sns.pairplot(iris, palette = "Set3")
```

```python
sns.pairplot(iris, hue = "species", palette = "Paired")
```

**LM plot - line plot in a scatter plot**

```python
sns.lmplot(x= "petal_length", y="petal_width", data =iris,  palette = 'tab10')
sns.lmplot(x= "petal_length", y="petal_width", data =iris, hue ="species")
```

**Markers (Symbols) for visual treat**




```python
sns.lmplot(x= "petal_length", y="petal_width", data =iris, hue ="species", markers = ["o","*","^"])
```

# **Box plot** 
**It generally represents the scatter values in the form of box in which the T shape line above & below the box represents the maximum & minimum peak value of the data while the center line inside the box reprsents the median(middle) value where more number of datas are present.**

**Note: The diamond shape reveals the error or negative value in the dataset that doesnot belong to any of the result.**
